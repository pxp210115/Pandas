Groupby is usually involved in one of the following use cases:-

* Splitting the data on specific( 1 or 2) conditions
* Applying a function to each group 

  * A summary statistic - Such as applying a sum/mean/median etc

  * Applying a unified operation on each group as needed

  * Replacing missing values specifically on the base of a grouping

  * Filtering on the Boolean values of True/False
  


## Aggregation Functions

```python
# Aplying the median function on the Age parameter of each of the groups
train.groupby('Pclass').Age.mean()
```

```python

train.groupby('Pclass').Fare.mean()
```

A thing to infer is that the average age of Pclass(1) or the highest ranking in the lot has an average fare which is 6.4 times the base price ticket for a Pclass(3) of $13.6

```python
train.groupby('Pclass').agg(['mean', 'median', 'sum'])
```

We can infer get mutlitple statistics at once using the agg ( [ ' passing mathematical functions as a list ' ] )

## Transformation functions

```python
# Transformation is often necassry for transforming the data into ML ready/ suitable numbers

# The use case is much more user friendly when dealing with Deep Learning models/ excessive data

standardization = lambda x: (x-x.mean())/ x.std()

# Does a standardization as per 
train.groupby('Pclass').transform(standardization)
```

```python
# Creates a new column and then assigns each groupby class the value of the sum of the fares of the entire group

# Unique values/ value counts in the 'total_fare_purchase' would be three unique values
train['total_fare_purchase']= train.groupby('Pclass')['Fare'].transform('sum')
train
```

## Grouping by Multiple categories

```python
# Splits the first argument over the various types of the second argument

# Can be used with aggregation features to get a statistic of the mixed grouping

train.groupby(['Survived', 'Pclass']).mean()
```

```python
# Can also reset the index of the grouping

# There is a catch here that it ignores all the Missing values
train.groupby(['Survived', 'Pclass']).mean()
```

```python
train.isna().sum()
```

```python
# To include the NA's and NAN as another group all together

df_subset = train.loc[:, ['Sex', 'Pclass', 'Age', 'Fare']]

# Creating the missing value

df_subset.iloc[80:100, 0] = np.nan

# Creating a third group 

df_subset.groupby(['Sex', 'Pclass'], dropna=False).mean()
```
